```python
combined_liar_df['label_num'] = 0
combined_liar_df.loc[combined_liar_df['label'] == 'pants-fire', 'label_num'] = 0 
combined_liar_df.loc[combined_liar_df['label'] == 'false','label_num'] = 1
combined_liar_df.loc[combined_liar_df['label'] == 'barely-true','label_num'] = 2
combined_liar_df.loc[combined_liar_df['label'] == 'half-true','label_num'] = 3
combined_liar_df.loc[combined_liar_df['label'] == 'mostly-true','label_num'] = 4
combined_liar_df.loc[combined_liar_df['label'] == 'true','label_num'] = 5
```

```python
def lst_to_str(lst):
    return " ".join(lst)
```

```python
def binary_class(label):
    if label in ('false','barely-true','pants-fire'):
        return 1
    elif label in ('half-true','mostly-true','true'):
        return 0
```

```python
def ternary_class(label):
    if label in ('mostly-true', 'true'):
        return 0
    elif label in ('half-true','barely-true'):
        return 1
    elif label in ('false','pants-fire'):
        return 2
```

```python
combined_liar_df['binary_label_num'] = combined_liar_df['label'].apply(binary_class)
combined_liar_df['ternary_label_num'] = combined_liar_df['label'].apply(ternary_class)
```

```python
import re

def remove_punctuation(tokens):
    """
    Remove all punctuation and special characters from a list of tokenized words.
    
    Parameters:
        tokens (list of str): List of tokenized words.
        
    Returns:
        list of str: List of tokenized words without punctuation and special characters.
    """
    pattern = r'[^\w\s]' # match any character that is not a word character or whitespace
    return [re.sub(pattern, '', token) for token in tokens if not re.match(pattern, token)]
```

```python
combined_liar_df['fear'] = 0.0
combined_liar_df['anger'] = 0.0
combined_liar_df['anticipation'] = 0.0
combined_liar_df['trust'] = 0.0
combined_liar_df['surprise'] = 0.0
combined_liar_df['positive'] = 0.0
combined_liar_df['negative'] = 0.0
combined_liar_df['sadness'] = 0.0
combined_liar_df['disgust'] = 0.0
combined_liar_df['joy'] = 0.0
```

```python
# create EmoLex features
count = 0
for index, row in combined_liar_df.iterrows():
    count+=1
    
    
        
    claim = row['combined_text'].strip()
    emotion = NRCLex(claim)
    emotion_dict = emotion.affect_frequencies
    
    for em in emotion_dict:
        if em == 'fear':
            combined_liar_df.iloc[index,7] = emotion_dict[em]
        if em == 'anger':
            combined_liar_df.iloc[index,8] = emotion_dict[em]
        if em == 'anticipation' or em == 'anticip':
            combined_liar_df.iloc[index,9] = emotion_dict[em]
        if em == 'trust':
            combined_liar_df.iloc[index,10] = emotion_dict[em]
        if em == 'surprise':
            combined_liar_df.iloc[index,11] = emotion_dict[em]
        if em == 'positive':
            combined_liar_df.iloc[index,12] = emotion_dict[em]
        if em == 'negative':
            combined_liar_df.iloc[index,13] = emotion_dict[em]
        if em == 'sadness':
            combined_liar_df.iloc[index,14] = emotion_dict[em]
        if em == 'disgust':
            combined_liar_df.iloc[index,15] = emotion_dict[em]
        if em == 'joy':
            combined_liar_df.iloc[index,16] = emotion_dict[em]
            

print(count)
```

```python
def pos_tagging(input_text):
    tokens = nltk.word_tokenize(input_text)
    pos_tags = nltk.pos_tag(tokens)

    # Combine the input text and POS tags using "#" as a delimiter
    input_with_pos = input_text + " #"
    for token, pos_tag in pos_tags:
        input_with_pos += pos_tag + " "

    # Remove the trailing whitespace
    input_with_pos = input_with_pos.strip()

    return input_with_pos
```

```python
LIWC_combined_df['combined_pos'] = LIWC_combined_df['combined_text'].apply(pos_tagging)
```
